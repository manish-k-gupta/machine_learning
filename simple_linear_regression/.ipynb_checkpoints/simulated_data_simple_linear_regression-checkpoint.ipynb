# Simple Linear Regression - With Simulated Data

```python
# to have matplotlib plots inline
%matplotlib inline
```

## Import Packages / Modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Data Simulation

1. Create uniformly distributed feature x with bounds of [0-100).
2. Create linearly generated output y with feature multiplied by a fixed weight and add random normal noise to it.

```python
array_size = 100
feature = np.random.uniform(0,100,array_size)
target = feature * 2 + 1 + np.random.normal(scale=5,size=array_size)
```

## Plot simulated data

```python
plt.hist(feature,10)
plt.xlabel("Features")
plt.show()
plt.scatter(feature,target)
plt.xlabel("Features")
plt.ylabel("Targets")
plt.show()
```

## Fit linear model
### Selected Model is single variable linear regression

$\hat{F(x)}$ = $\hat{W0}$ + $\hat{W1}$X

$\hat{Y}$ = $\hat{W0}$ + $\hat{W1}$X (as we dont know whether the observed is above line or below line, we put it on line itself.)

$Y_i$ = $W0_i$ + $W1_i$X + $\epsilon$

W0 & W1 are regression coefficient & $\epsilon$ is estimation / observation error. 

```python
A,B = np.random.rand(2)
estimate = A * feature  + B
print("slope => %s" %A)
print("intercept => %s" %B)
```

## Plot target verses estimate
1. Red color line - estimate
2. Blue color line - target 

```python
plt.scatter(feature,estimate,color='r')
plt.scatter(feature,target,color='b')
plt.xlabel("feature")
plt.ylabel("target / estimate")
plt.show()
```

## ALGORITHM
## ------------------   

### Cost function is Residual Sum of Squares which is the sum of square of differences between observed and estimated value of target.

### Evaluate A & B

```python
denominator = feature.dot(feature) - feature.mean()*feature.sum()
A = (feature.dot(target) - target.mean()*feature.sum()) / denominator
B = (target.mean()*feature.dot(feature) - feature.mean()*feature.dot(target)) / denominator
print("slope => %s" %A)
print("intercept => %s" %B)
```

```python
estimate = A * feature + B
```

## Plot newly calculated estimate

```python
plt.scatter(sorted(feature),sorted(estimate),color='r')
plt.scatter(sorted(feature),sorted(target),color='b')
plt.show()
```

## Determine the strength of model by computing R-Squared.

```python
diff_target_estimate = target - estimate
diff_target_targetmean = target - target.mean()
R2 = 1 - (diff_target_estimate.dot(diff_target_estimate) / diff_target_targetmean.dot(diff_target_targetmean))
print(R2)
```
